```python


import pandas as pd

# Webpage url                                                                                                               
url = 'https://en.wikipedia.org/wiki/50_Greatest_Players_in_NBA_History'

# Extract tables
dfs = pd.read_html(url)

# Get first table                                                                                                           
df = dfs[3]

# Extract columns                                                                                                           
df = df[['Name','Pos']]
print(df)
```

```python
# Write to excel
df.to_excel('NBA.xlsx')
```
